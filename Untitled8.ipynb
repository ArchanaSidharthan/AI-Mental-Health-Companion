```python
# Install dependencies
!pip install -q transformers torchaudio gradio

import torch
import torchaudio
import gradio as gr
from transformers import AutoModelForAudioClassification, AutoFeatureExtractor

# Load pretrained model for Emotion Recognition
MODEL_NAME = "superb/wav2vec2-base-superb-er"
model = AutoModelForAudioClassification.from_pretrained(MODEL_NAME)
extractor = AutoFeatureExtractor.from_pretrained(MODEL_NAME)

# Label map (emotions that this model knows)
id2label = model.config.id2label

# Prediction function
def predict_emotion(audio_file):
    # Load and resample audio
    speech_array, sr = torchaudio.load(audio_file)
    if sr != 16000:
        resampler = torchaudio.transforms.Resample(sr, 16000)
        speech_array = resampler(speech_array)
    # Extract features
    inputs = extractor(speech_array.squeeze().numpy(), sampling_rate=16000, return_tensors="pt")
    with torch.no_grad():
        logits = model(**inputs).logits
    pred = torch.argmax(logits, dim=-1).item()
    label = id2label[pred]

    # Emoji mapping (you can adjust if needed)
    emoji_map = {
        "angry": "😡 Angry",
        "happy": "🙂 Happy",
        "sad": "😢 Sad",
        "neutral": "😐 Neutral",
        "fear": "😱 Fear",
        "disgust": "🤢 Disgust",
        "surprise": "😮 Surprise",
    }
    return emoji_map.get(label, label)

# Gradio interface
iface = gr.Interface(
    fn=predict_emotion,
    inputs=gr.Audio(type="filepath", label="Upload or Record your voice"),
    outputs=gr.Textbox(label="Predicted Emotion"),
    title="🎤 Pretrained Speech Emotion Recognition",
    description="Upload or record audio and get the predicted emotion instantly."
)

iface.launch(debug=True)
```
